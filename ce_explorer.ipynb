# CE 2.0 Explorer

```python
from ce import *
from ce_systems import *
```

```python
def explore(tpm):
    path = createPath(tpm)
    
    for step in path:
        tpm_rounded = np.round(step[0], 4)
        print(f"TPM:\n{tpm_rounded}\nCP: {step[1]:.4f}, Determinism: {step[2][0]:.4f}, Specificity: {step[2][1]:.4f}, Merged Pair: {step[3]}\n")

    changeplot = plot_cp_change(path)
    absplot = plot_absolute_scores(path)

    return (changeplot, absplot)
```

```python
changeplot, absplot = explore(rand_8x8) # for example
changeplot.show()
absplot.show()
```
